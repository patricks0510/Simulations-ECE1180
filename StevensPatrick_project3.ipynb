# Assumptions and Setup
Each node is attached to those adjacent to it
all nodes have the same mass, in kilograms
all springs are identical
spacing in meters
scaled up for simulation purposes, easier than dealing with decimals or exponentials

```python
#IMPORTS and VPython Settings
from vpython import *
import numpy as np
import math
# importing matplotlib module
from matplotlib import pyplot as plt
scene=canvas()
autoscale = False
```

```python
#Simulation Constants and variables
#dimension of matrix
N = 3
#spring constant [N/m] or [kg/s^2]
k = 1
#mass of nodes [kg]
m = 1

t = 0
dt = 0.05
freq = 1/dt
endTime = 6

#time in minutes and seconds for displaying
t_m = 0
t_s = t
```

```python
#Object Creation Block

#spacing between nodes and relative radius [m]
spacing = 1
#radius for visiualization [m]
node_radius = 0.3*spacing
#length of springs between nodes [m]
L0 = spacing-1.8*node_radius
# initial volume of spring [m^3]
V0 = math.pi*(0.5*node_radius)**2*L0

#set scene centered on center of visible nodes
scene.center = 0.5*(N-1)*vector(1,1,1)
#set background to white
scene.background = color.white
#define axes of space with unit vectors
axes = [vector(1,0,0), vector(0,1,0), vector(0,0,1)]

class crystal:
    
    def __init__(self,N,node_radius,spacing,velocityRange):
        #nodes and springs dictionaries
        self.nodes = []
        self.springs = []

        #create matrix of nodes, -1 to N+1 to make invisible stablizer nodes
        for z in range(-1,N+1,1):
            for y in range(-1,N+1,1):
                for x in range(-1,N+1,1):
                    #define node and position
                    node = sphere()
                    node.pos = vector(x,y,z)*spacing
                    node.radius = node_radius
                    node.color = vector(0.4,0.2,0.6)
                    if 0<=x<N and 0<=y<N and 0<=z<N:
                        #randomized initial vector for part c
                        direction = vec.random()
                        #velocity times velocity range ot give directed velocity to node
                        node.velocity = velocityRange*direction
                    else:
                        #invisible stablizing nodes surrounding the 3x3 simulation
                        node.visible = False
                        node.velocity = vector(0,0,0)
                    #add node to matrix dictionaries with index
                    node.index = len(self.nodes)
                    self.nodes.append( node )
        #connect nodes with springs
        for node in self.nodes:
            #only visible nodes, not surrounding stablizing nodes
            if node.visible:
                if node.pos.x == 0:
                    #make invisible spring to previous invisible node
                    self.make_spring(self.nodes[node.index-1], node, False)
                    #make visible spring to next visible node in x direction
                    self.make_spring(node, self.nodes[node.index+1], True)
                elif node.pos.x == N-1:
                    #make invisible spring to next node when at N-1
                    self.make_spring(node, self.nodes[node.index+1], False)
                else:
                    #make visible spring when in the middle
                    self.make_spring(node, self.nodes[node.index+1], True)

                if node.pos.y == 0:
                    self.make_spring(self.nodes[node.index-(N+2)], node, False)
                    self.make_spring(node, self.nodes[node.index+(N+2)], True)
                elif node.pos.y == N-1:
                    self.make_spring(node, self.nodes[node.index+(N+2)], False)
                else:
                    self.make_spring(node, self.nodes[node.index+(N+2)], True)
                    
                if node.pos.z == 0:
                    self.make_spring(self.nodes[node.index-(N+2)**2], node, False)
                    self.make_spring(node, self.nodes[node.index+(N+2)**2], True)
                elif node.pos.z == N-1:
                    self.make_spring(node, self.nodes[node.index+(N+2)**2], False)
                else:
                    self.make_spring(node, self.nodes[node.index+(N+2)**2], True)
                    
                     
    # Create a grid of springs linking each node to the adjacent nodes
    # in each dimension, or to invisible motionless nodes
    def make_spring(self, start, end, visible):
        spring = helix()
        spring.pos = start.pos
        spring.axis = end.pos-start.pos
        spring.visible = visible
        spring.thickness = 0.05
        spring.radius = 0.5*node_radius
        spring.length = spacing
        #starting and ending nodes of the spring
        spring.start = start
        spring.end = end
        spring.color = color.orange
        self.springs.append(spring)
        
#crystal with NxN visible nodes
# 0.1*spacing*sqrt(k/m) in units [m/s] so it is a velocity
c = crystal(N, node_radius, spacing, 0.1*spacing*sqrt(k/m))

#get system energy
# KE = 1/2 * momentum * velocity
# PE = 1/2 * spring constant * displacement^2
pEnergy = 0
kEnergy = 0
for x in c.nodes:
    kEnergy += 0.5*m*(mag(x.velocity)**2)
# for x in c.springs:
#     pEnergy += 0.5*k*((1-spacing/L0)**2)

sysEnergy = pEnergy + kEnergy
pEnergies = []
kEnergies = []
sysEnergies = []
tArray = []
meetingPoint = []
meetingTimes = []

pEnergies.append(pEnergy)
kEnergies.append(kEnergy)
sysEnergies.append(sysEnergy)
tArray.append(t)
print(pEnergy)
print(kEnergy)
print(sysEnergy)
print("************")
```

# Forward Euler

```python
#Simulation Loop
while t < endTime:
    rate(freq)
    #reset kinetic and potential energy
    kEnergy = 0
    pEnergy = 0
    sysEnergy = 0
    for node in c.nodes:
        if node.visible:
            #update position of node
            #forward Euler uses previous position and previous velocity to get next position
            node.pos = node.pos + (node.velocity)*dt
            #add up new kinetic energy
            kEnergy += 0.5*m*(mag(node.velocity)**2)
    for spring in c.springs:
        spring.axis = spring.end.pos - spring.start.pos
        #new length of spring
        L = mag(spring.axis)
        #axis of spring as unit vector
        spring.axis = spring.axis.norm()
        #move spring position to new location depending on stretch
        spring.pos = spring.start.pos+0.5*node_radius*spring.axis
        #take out node radius from spring length
        spring.length = L-node_radius
        # Potential Energy calc
        pEnergy += 0.5*k*((1-spacing/L)**2)
        # [vector]*[kg/s^2]*[s]*[] = vector [kg/s]
        Fdt = spring.axis * (k*dt * (1-spacing/L))
        #apply spring force through timestep to change node velocity
        #get next velocity with current velocity and currecnt force/accel, forward euler
        if spring.start.visible:
            spring.start.velocity += Fdt
        if spring.end.visible:
            spring.end.velocity -=  Fdt
    #plotting energies later
    sysEnergy = kEnergy + pEnergy
    pEnergies.append(pEnergy)
    kEnergies.append(kEnergy)
    sysEnergies.append(sysEnergy)
    #update the caption with the current elapsed time
    t+=dt
    tArray.append(t)
    
    if(kEnergy == pEnergy):
        meetingPoint.append(kEnergy)
        print(t)
    
#     print(pEnergy)
#     print(kEnergy)
#     print(sysEnergy)
#     print("************")
    t_m,t_s = divmod(t,60)
    scene.caption = "Time Elapsed: {}:{:.3f}".format(int(t_m),round(t_s,3))
```

```python
plt.plot(tArray, pEnergies)
plt.plot(tArray, kEnergies)
plt.plot(tArray,sysEnergies)
plt.legend(["potential","kinetic","total"])
plt.show()
```

# Analysis of Forward Euler

As you can see, when potential energy rises, kinetic energy falls and vice versa. The drastic jump from the initial values is dues to the fact that the Forward Euler uses no information about next steps to calculate the next values

# Backward Euler

```python
scene = canvas()
#set scene centered on center of visible nodes
scene.center = 0.5*(N-1)*vector(1,1,1)
#set background to white
scene.background = color.white

c2 = crystal(N, node_radius, spacing, 0.1*spacing*sqrt(k/m))

#get system energy
# KE = 1/2 * momentum * velocity
# PE = 1/2 * spring constant * displacement^2
t=0
pe2 = 0
ke2 = 0
for x in c2.nodes:
    ke2 += 0.5*m*(mag(x.velocity)**2)
# for x in c.springs:
#     pEnergy += 0.5*k*((1-spacing/L0)**2)

se2 = pe2 + ke2
pe2s = []
ke2s = []
se2s = []
t2s = []
meetingPoint2 = []
meetingTimes2 = []
pe2s.append(pe2)
ke2s.append(ke2)
se2s.append(se2)
t2s.append(t)


#Simulation Loop
while t < endTime:
    rate(freq)
    #reset kinetic and potential energy
    ke2 = 0
    pe2 = 0
    se2 = 0
    
    for spring in c2.springs:
        spring.axis = spring.end.pos - spring.start.pos
        #new length of spring
        L = mag(spring.axis)
        #axis of spring as unit vector
        spring.axis = spring.axis.norm()
        #move spring position to new location depending on stretch
        spring.pos = spring.start.pos+0.5*node_radius*spring.axis
        #take out node radius from spring length
        spring.length = L-node_radius
        # Potential Energy calc
        pe2 += 0.5*k*((1-spacing/L)**2)
        # [vector]*[kg/s^2]*[s]*[] = vector [kg/s]
        Fdt = spring.axis * (k*dt * (1-spacing/L))
        #apply spring force through timestep to change node velocity
        #since velocity is changing before node position, next velocity is used for integration, backward euler
        if spring.start.visible:
            spring.start.velocity += Fdt
        if spring.end.visible:
            spring.end.velocity -=  Fdt
    for node in c2.nodes:
        if node.visible:
            #update position of node
            #backward Euler uses current position and next velocity for next position
            node.pos = node.pos + (node.velocity)*dt
            #add up new kinetic energy
            ke2 += 0.5*m*(mag(node.velocity)**2)
    
    #plotting energies later
    se2 = ke2 + pe2
    pe2s.append(pe2)
    ke2s.append(ke2)
    se2s.append(se2)
    #update the caption with the current elapsed time
    t+=dt
    t2s.append(t)
    
    if(ke2 == pe2):
        meetingPoint.append(ke2)
        print(t)
    
#     print(pEnergy)
#     print(kEnergy)
#     print(sysEnergy)
#     print("************")
    t_m,t_s = divmod(t,60)
    scene.caption = "Time Elapsed: {}:{:.3f}".format(int(t_m),round(t_s,3))
```

```python
plt.plot(t2s, pe2s)
plt.plot(t2s, ke2s)
plt.plot(t2s,se2s)
plt.legend(["potential","kinetic","total"])
plt.show()
```

# Analysis of Backward Euler
Kinetic and Potential energies are still inversely related which is good, but the initial changes are not as drastic due to the fact that the backwards eutler uses the next calculated velocity rather than the current one. This "premonition" allows for a smoother integration

# Trapezoid

```python
scene = canvas()
#set scene centered on center of visible nodes
scene.center = 0.5*(N-1)*vector(1,1,1)
#set background to white
scene.background = color.white

c3 = crystal(N, node_radius, spacing, 0.1*spacing*sqrt(k/m))

#get system energy
# KE = 1/2 * momentum * velocity
# PE = 1/2 * spring constant * displacement^2
t=0
pe3 = 0
ke3 = 0
for x in c3.nodes:
    ke3 += 0.5*m*(mag(x.velocity)**2)
# for x in c.springs:
#     pEnergy += 0.5*k*((1-spacing/L0)**2)

se3 = pe3 + ke3
pe3s = []
ke3s = []
se3s = []
t3s = []
meetingPoint3 = []
meetingTimes3 = []
pe3s.append(pe3)
ke3s.append(ke3)
se3s.append(se3)
t3s.append(t)

#Simulation Loop
while t < endTime:
    rate(freq)
    #reset kinetic and potential energy
    ke3 = 0
    pe3 = 0
    se3 = 0
    #current velos used for trapezoid integrations
    currentVelos = []
    for node in c3.nodes:
        currentVelos.append(node.velocity)
    for spring in c3.springs:
        spring.axis = spring.end.pos - spring.start.pos
        #new length of spring
        L = mag(spring.axis)
        #axis of spring as unit vector
        spring.axis = spring.axis.norm()
        #move spring position to new location depending on stretch
        spring.pos = spring.start.pos+0.5*node_radius*spring.axis
        #take out node radius from spring length
        spring.length = L-node_radius
        # Potential Energy calc
        pe3 += 0.5*k*((1-spacing/L)**2)
        # [vector]*[kg/s^2]*[s]*[] = vector [kg/s]
        Fdt = spring.axis * (k*dt * (1-spacing/L))
        #apply spring force through timestep to change node velocity
        #since velocity is changing before node position, next velocity is used for integration
        #current velocity stored in its own list
        if spring.start.visible:
            spring.start.velocity += Fdt
        if spring.end.visible:
            spring.end.velocity -= Fdt
    for idx, node in enumerate(c3.nodes):
        if node.visible:
            #update position of node
            #trapezoid uses both next velocity and current velocity averaged
            node.pos = node.pos + (node.velocity+currentVelos[idx])/2 *dt
            #add up new kinetic energy
            ke3 += 0.5*m*(mag(node.velocity)+mag(currentVelos[idx]))/2 **2
    
    #plotting energies later
    se3 = ke3 + pe3
    pe3s.append(pe3)
    ke3s.append(ke3)
    se3s.append(se3)
    #update the caption with the current elapsed time
    t+=dt
    t3s.append(t)
    
    if(ke3 == pe3):
        meetingPoint3.append(ke3)
        print(t)
    
#     print(pEnergy)
#     print(kEnergy)
#     print(sysEnergy)
#     print("************")
    t_m,t_s = divmod(t,60)
    scene.caption = "Time Elapsed: {}:{:.3f}".format(int(t_m),round(t_s,3))
```

```python
plt.plot(t3s, pe3s)
plt.plot(t3s, ke3s)
plt.plot(t3s,se3s)
plt.legend(["potential","kinetic","total"])
plt.show()
```
