# RC Circuit
To create the RC circuit, rather than focusing on the individual components and connecting them with a netlist (something I struggled with for the logic networks project), I took the advice from Caleb Troyer to create a function that does all the math for the whole circuit at once, rather than passing data back and forth between component objects/arrays. This simplified the process of calculating Vout.

```python
#imports
import vpython
import math
from matplotlib import pyplot as plt
import numpy as np
```

```python
#Global Constants
TSTEP = 5.0e-6 # [s]

ENDTIME = 1.0 # [s]

VOLT0 = 5 #[V]

AC_FREQ = 40 #[Hz]

C = 1.0e-5 #[F]

R = 1000.0 #[ohms]

TAU = (R*C)
```

```python
def rc_stepper(R,C,Vi,Ieq):
    
    matA = np.array([[-1.0/R,1.0/R,1.0],
                    [1.0/R,(-1.0/R)-(C/TSTEP),0],
                    [1.0,0,0]])
    matB = np.array([0, Ieq, Vi])
    
    matX = np.linalg.solve(matA,matB)
    
    return matX[1]
```

```python
#sim loop
currTime = 0
#initialize vout with value for first calcualtion
Vout = [1.0/TAU]
#simulation loop
while currTime<ENDTIME:
    #calculate Vi = V0*sin(2*math.pi*vFreq*t)
    #or vi = 0 for impulse response
    Vi = 0 #VOLT0*math.sin(2*math.pi*AC_FREQ*currTime)
    Ieq = -1*C/TSTEP*Vout[-1]
    Vout.append(rc_stepper(R,C,Vi,Ieq))
    currTime += TSTEP
    
timeArr = np.arange(0.0,ENDTIME,TSTEP)
#remove initial value of vout to keep timeArr and Vout same size
Vout = Vout[1:]

#helpful function for comparing the expected impulse response, collaborated with Caleb Troyer for this
def expected_impulse_response(TAU,ts):
    return [
        1/TAU * math.exp(-t/TAU) if t>= 0 else 0.0
        for t in ts
    ]

expectedImpulse = expected_impulse_response(TAU,timeArr)
# plot
plt.style.use('_mpl-gallery')
fig, ax = plt.subplots(2,1,figsize=(10,10))

ax[0].plot(timeArr, Vout, label = "Simulated Impulse Response", linewidth=1.0)

ax[0].set(xlim=(0, .1), xticks=np.arange(0, .1, .01),
       ylim=(-1, 100), yticks=np.arange(-1, 100,10))

ax[1].plot(timeArr,expectedImpulse,label = "Mathematical Imulse Response",linewidth=1.0)

ax[1].set(xlim=(0, .1), xticks=np.arange(0, .1, .01),
       ylim=(-1, 100), yticks=np.arange(-1, 100,10))

plt.legend()


plt.show()
```

I chose these values as they give an impulse response that takes under 1 second to settle down to 0V on the output.

As you can see, these plots are essentially identical, at least visually. There may be an error, but the error factor is less than .1%

# Half Wave Rectifier
Again going with a procedural programming approach rather than a component/netlist OOP approach, I used np arrays of a certain size before the simulation loop. When I had empty arrays and appended them for the RC ciruit, I ended up with some off by one errors and instead used the numpy function zeros_like, which creates an array of zeros the same length as an imput array. Since I had the size of my arrays already determined, I instead used a for-loop to reference the indeces of the np arrays, rather than appending with a while-loop.

```python
#circuit constants
R = 1000 #[ohms]
C = 1e-7 #[farads]
TAU = R*C #time constant that depends on resistance and capacitance
VOLT0 = 5 #[volts]
AC_FREQ = 100 #[Hz]

#simulation constants
TSTEP = 1e-6
ENDTIME = 0.1

#time and input voltage arrays for simulation
timeArr = np.arange(0,ENDTIME,TSTEP)
#using np.sin rather than math.sin to get an array like object as a return
voltInArr = VOLT0*np.sin(2*np.pi*AC_FREQ*timeArr)

#starting output voltage and diode voltage
Vout = np.zeros_like(voltInArr)
Vdiode = np.zeros_like(voltInArr)

#sim loop here
for i in range(1,len(timeArr)):
    #diode voltage calc
    Vdiode[i] = voltInArr[i]-Vout[i-1]
    
    #convergence check
    for j in range(10):
        prevVdiode = Vdiode[i]
        Vdiode[i] = voltInArr[i]-Vout[i-1] + (VOLT0/math.pi)*math.sin(2*math.pi*AC_FREQ*timeArr[i])
        if abs(Vdiode[i]-prevVdiode)<TSTEP:
            break
            
    #overwrite zeros in vout
    Vout[i] = Vout[i-1]+TSTEP*(voltInArr[i]-Vout[i-1])/TAU * (Vdiode[i] >=0)
    
#dc component as the mean of the output
Vdc = np.mean(Vout)

#plotting
plt.plot(timeArr,voltInArr,label = "V_in")
plt.plot(timeArr,Vout,color="red",label = "V_out")
plt.title("half wave rectifier")
plt.xlabel("time [s]")
plt.ylabel("voltage [V]")
plt.legend()
plt.show()

print("DC part of output",Vdc)
```

# Analysis
The input voltage in blue varies between 5 and -5, but the output voltage, rectified for a half wave, never drops below 0. As R or C is increased, so does the value of the bottom of the half wave, which is expected and accurate to the physical properies of a Half Wave Rectifier.
